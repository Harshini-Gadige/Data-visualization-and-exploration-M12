```python
# Determine the five most common journals and the total articles for each.
import pandas as pd
```

```python
data = pd.read_csv("/Users/harshini/Desktop/Thinkful/Data/book.csv", encoding = "ISO-8859-1")
data.head()
```

```python
top5_list = list(data['Journal title'].value_counts().index[0:5])
top5_list
```

```python
group_jour = data.groupby('Journal title').count()
group_jour.head()
```

```python
top = group_jour.loc[top5_list,'Article title']
top_df = pd.DataFrame()
top_df['Top 5 Journals'] = top.index
top_df['Article count']=list(top)
top_df
```
